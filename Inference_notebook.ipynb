# Esta es la aplicacion para poder hacer inferencia de nuestro modelo serializado

```python
# CELDA 1: cargar el pipeline entrenado (joblib) y crear un ejemplo hardcodeado
import joblib
import numpy as np

model = joblib.load("best_svc_pipeline.joblib")  # ajusta ruta si esta en otra carpeta

# ejemplo hardcodeado (4 features) -> CAMBIA estos valores por unos reales si quieres
x_new = np.array([[1.2, 85.0, 28.0, 0.35]])  # shape (1,4)
print("Input:", x_new)
```

```python
# CELDA 2: inferencia (prediccion) usando el modelo cargado
y_pred = model.predict(x_new)

# si quieres mostrar score (SVC tiene decision_function)
score = model.decision_function(x_new)

print("Prediccion (0=no diabetes, 1=diabetes), prediccion real:", int(y_pred[0]))
```
